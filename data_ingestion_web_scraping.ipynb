```python
import requests
import pandas as pd
import numpy as np
import altair as alt
from bs4 import BeautifulSoup
import re
```

## Bharath Kalachar

```python
URL = "https://www.bharatkalachar.com/34th-margazhi-mahothsav.html"
page = requests.get(URL)
```

```python
soup = BeautifulSoup(page.content, "html.parser")
table = soup.find_all("table")
```

```python
rows = []
for child in soup.find_all('table')[0].children:
    row = []
    for td in child:
        try:
            row.append(td.text)
        except:
            continue
    if len(row) > 0:
        rows.append(row)

df = pd.DataFrame(rows[1:], columns=rows[0])


df1 = pd.DataFrame()
for i in range(7):
    df2 = pd.DataFrame()
    df2['Name'] = df['NAME'].str.split("\r\n", expand = True)[i]
    df2['Instrument'] = df['SPECIALITY'].str.split("\r\n", expand = True)[i]
    df1 = df1.append(df2)
df = df.merge(df1, how = 'outer', right_index = True, left_index = True )

df.DATE = df.DATE.str.strip()
df[['Date', 'Month', 'Year']] = df.DATE.str.split(".", expand = True)
df.Date = pd.to_numeric(df.Date)
df.Month = pd.to_numeric(df.Month)
df.Year = pd.to_numeric(df.Year)

df[['Date', 'Month', 'Year']] = df[['Date', 'Month', 'Year']].ffill()
df.TIME = df.TIME.str.replace('.', ":")
df.TIME = df.TIME.str.replace(',', ":")
df.TIME = df.TIME.str.replace('a', "A")
df.TIME = df.TIME.str.replace('p', "P")
df.TIME = df.TIME.str.replace('m', "M")
df.TIME = df.TIME.str.replace(" ", "")
df['Date'] = df.Date.astype(int).astype(str)
df['Month'] = df.Month.astype(int).astype(str)
df['Year'] = df.Year.astype(int).astype(str)
df['startDateTime'] = pd.to_datetime(df.Year + "-" + df.Month + "-" +  df.Date + " " + df.TIME)

df['Name'] = df.Name.astype(str)
df['Instrument'] = df.Instrument.astype(str)
df['sabha'] = "bharatkalachar"
df['address'] = "No.16.Thirumalai Pillai Street, T-Nagar, 17, Thirumalai Pillai Rd, T. Nagar, Chennai, Tamil Nadu 600017, India"
df['lat'] = 13.049258196982954
df['long'] = 80.2410428821105

df.Name = df.Name.str.replace("\n", "")
df.Name = df.Name.str.strip()
df.Name = df.Name.str.replace('    ', "nan")
df.Instrument = df.Instrument.str.strip()

df['Instrument'] = df.Instrument.str.lower()
df.Instrument = df.Instrument.str.replace("\t", "")
df.Instrument = df.Instrument.str.replace("\n", "")
df = df.query('Name != "nan"')
```
